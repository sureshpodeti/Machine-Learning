```python
# Load the boston housing prices dataset
# do regresssion using gradient descent algorithm
import numpy as np
from sklearn import datasets, model_selection, preprocessing, linear_model
import matplotlib.pyplot as plt
```

```python
# Hypothesis: LINEAR : h(x) = theta0*x0 + theta*x1 + ...., where x0=
```

```python
# Load the dataset
boston = datasets.load_boston()
x = boston.data
y = boston.target
```

```python
### Data preprocessing

## feature scaling/ mean normalization
x = preprocessing.scale(x)

## splitting dataset into train,test 
x_train, x_test, y_train ,y_test = model_selection.train_test_split(x,y, test_size=0.3, random_state=42)
m, n = x_train.shape[0], len(boston.feature_names)
# constructing A, b arrays
ones = np.ones((m,1))
A = np.append(ones, x_train, axis=1)
b = y_train
```

```python
def learn_params(theta, A, b, m,n, alpha=0.01):
    for i in range(0,n+1):
        value = 0
        for j in range(0,m):
            value += (np.dot(theta.T, np.reshape(A[j:j+1,], (n+1,1)))[0,0]-b[j])*A[j,i]
        theta[i,0] -= alpha*value
    return theta
            
```

```python
# Initialize theta vector to all zeros
theta = np.zeros((n+1,1)) # theta = [theta0, theta1, ...thetan]
# define learning rate 
alpha = 0.01
copy_theta = np.copy(theta)

while True:
    new_theta = learn_params(np.copy(theta), A, b, m,n, alpha)
    # convergence condition is sum of squrred errors should be minimum
    J = sum((np.dot(A, theta) - np.reshape(b, (m,1)))**2)[0]
    e = sum((np.dot(A, new_theta) - np.reshape(b,(m,1)))**2)[0]
    #print(J, e)
    #print(new_theta)
    
    if abs(J-e)<=0.001:
        print(new_theta)
        break
    theta = np.copy(new_theta)
        
```
